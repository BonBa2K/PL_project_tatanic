```python
#常用套件
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#機器學習套件組
from sklearn import tree
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold, learning_curve, train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import RFECV
```

```python
pc = [1,2,3]
se = [0,1]
ag = [0,1]
ca = ['A','B','C','D','E','F','G']
em = ['S','C','Q']
fn = [0,1,2,3,4]
```

```python
t_data = {'PassengerId':[],'Survived':[],'Pclass':[],'Name':[],'Sex':[],'Age':[],
          'SibSp':[],'Parch':[],'Ticket':[],'Fare':[],'Cabin':[],'Embarked':[],
          'Sex_Code':[],'A16':[],'Cnum':[],'Fnum':[]}
#surv.append(a,ignore_index=True)
surv = pd.DataFrame(t_data)
surv
```

```python
for p in pc:
    for s in se:
        for a in ag:
            for c in ca:
                for e in em:
                    for f in fn:
                        #print(p,s,a,x,y,c,e)
                        re = {'PassengerId':'','Survived':'','Pclass':p,'Name':'','Sex':'','Age':'',
                              'SibSp':'','Parch':'','Ticket':'','Fare':'','Cabin':'','Embarked':e,
                              'Sex_Code':s,'A16':a,'Cnum':c,'Fnum':f}
                        surv = surv.append(re,ignore_index=True)
surv
```

```python
surv['Cbcde'] = surv['Cnum'].apply(lambda x: 1 if x in ['B','C','D','E'] else 0)
```

```python
#產出存活對照表基底
surv.to_csv("survival_table_base.csv",index=False)
```
